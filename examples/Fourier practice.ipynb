```python
# Import libraries
import numpy as np
from scipy.fftpack import fft, fftshift
from matplotlib import pyplot as plt

# Determine number of samples and sample rate
nsamples = 6000
srate = 800 # 5 Hz

# Determine time step (i.e. sample period)
tstep = 1/srate

# Create timestream
t = np.linspace(0, tstep*nsamples, nsamples)

# Create a superposition of sine signals
y0 = np.sin(2*np.pi*t*10)
y1 = np.sin(2*np.pi*t*20)
y2 = np.sin(2*np.pi*t*30)
y3 = np.sin(2*np.pi*t*40)
y = y0 + y1 + y2 + y3

# Plot
plt.plot(t, y0, label='y0')
plt.plot(t, y1, label='y1')
#plt.plot(t, y2, label='y2')
#plt.plot(t, y3, label='y3')
plt.plot(t, y, label='y')
plt.legend(loc=1)
plt.show()
```

```python
# Take FFT
fy = np.abs(fft(y))
fy = fy[0:int(len(fy)/2)]

fx = np.linspace(0, 1/(2*tstep), int(nsamples/2))

plt.plot(fx, fy)
plt.xlim(0,60)
plt.show()
```
